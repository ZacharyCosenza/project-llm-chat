```python
from evaluate import load

# === PERPLEXITY ===
perplexity = load("perplexity", module_type="metric")
results = perplexity.compute(
    predictions=["The cat sat on the mat"],
    model_id="gpt2"  # Uses GPT-2 to compute perplexity
)
print(f"Perplexity: {results['mean_perplexity']:.2f}")

# === BLEU (translation quality) ===
bleu = load("bleu")
results = bleu.compute(
    predictions=["the cat is on the mat"],
    references=[["the cat sat on the mat"]]  # Can have multiple references
)
print(f"BLEU: {results['bleu']:.4f}")

# === ROUGE (summarization) ===
rouge = load("rouge")
results = rouge.compute(
    predictions=["The cat sat on the mat"],
    references=["A cat was sitting on a mat"]
)
print(f"ROUGE-1: {results['rouge1']:.4f}")
print(f"ROUGE-2: {results['rouge2']:.4f}")
print(f"ROUGE-L: {results['rougeL']:.4f}")

# === Accuracy ===
accuracy = load("accuracy")
results = accuracy.compute(
    predictions=[0, 1, 2, 3, 1],
    references=[0, 1, 2, 2, 1]
)
print(f"Accuracy: {results['accuracy']:.2%}")

# === F1 Score (precision + recall) ===
f1 = load("f1")
results = f1.compute(
    predictions=[0, 1, 0, 1, 1],
    references=[0, 1, 1, 1, 0],
    average="binary"
)
print(f"F1: {results['f1']:.4f}")

# === METEOR (better than BLEU for some tasks) ===
meteor = load("meteor")
results = meteor.compute(
    predictions=["the cat is on the mat"],
    references=["the cat sat on the mat"]
)
print(f"METEOR: {results['meteor']:.4f}")

# === BERTScore (semantic similarity) ===
bertscore = load("bertscore")
results = bertscore.compute(
    predictions=["the cat is on the mat"],
    references=["a feline sits on a rug"],
    lang="en"
)
print(f"BERTScore F1: {results['f1'][0]:.4f}")
```
