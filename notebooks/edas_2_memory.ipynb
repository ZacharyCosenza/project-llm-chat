```python
import torch
import torchvision.models as models

# Load a pre-trained ResNet18 model
model = models.resnet18(weights='DEFAULT')
```

```python
model
```

```python
# Number of bytes in parameters
params_mb = sum([{'float32': 4, 'float16': 2, 'bfloat16': 2}[str(p.dtype).split('.')[-1]] for p in model.parameters()]) / 1e6
print(params_mb)

# Number of bytes in optimization
grads_mb = params_mb
optimizer_mb = params_mb * 2  # Adam: momentum + variance
print(grads_mb)
print(optimizer_mb)

# Number of btypes for activation
sample_input = torch.rand((1, 3, 64, 64))
torch.cuda.reset_peak_memory_stats()
with torch.no_grad():
    _ = model(sample_input)
activations_mb = torch.cuda.max_memory_allocated() / 1e6
```
